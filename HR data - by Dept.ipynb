```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
os.getcwd()
```

```python
hr=pd.read_csv('hr_data.txt')
hr['dept']=hr['sales']
hr.drop('sales',1,inplace=True)
```

```python
q={'low':0,'medium':1,'high':2}
hr['salary_num']=[q[i] for i in hr['salary']]
```

```python
hr_dept=hr.groupby('dept').mean()
```

```python
hr_dept.head(3)
```

```python
hr_dept1=pd.DataFrame(hr_dept['satisfaction_level'].sort_values(ascending=False),index=hr_dept.index)
hr_dept2=pd.DataFrame(hr_dept['left'].sort_values(ascending=True),index=hr_dept.index)
hr_dept3=pd.DataFrame(hr_dept['salary_num'].sort_values(ascending=True),index=hr_dept.index)
hr_dept.head(4)
```

```python
print(hr_dept.corr().mean().sort_values(ascending=False))
```

```python
f, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(hr_dept.corr(),linewidths=0.5,vmax=1,square=True,annot=True,cmap='Set3')
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,3,1)
sns.heatmap(hr_dept1,cmap="YlGnBu",annot=True,linecolor='black',xticklabels='auto', yticklabels='auto')
plt.subplot(1,3,2)
sns.heatmap(hr_dept3,annot=True,cmap="YlGnBu")
plt.subplot(1,3,3)
sns.heatmap(hr_dept2,annot=True,cmap="YlGnBu")
plt.tight_layout()
```
